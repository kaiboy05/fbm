```python
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
```

```python
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

from fbm.sim import naive
from fbm.testing.chi_square import fBm_chi_square_test
```

```python
size = 1000
fbm_gen = naive.NaiveFBmGenerator()
h05 = fbm_gen.generate_fBm(H=0.5, size=size)
h01 = fbm_gen.generate_fBm(H=0.1, size=size)

plt.plot(range(size), h05, label='H=0.5')
plt.plot(range(size), h01, label='H=0.1')
plt.legend()
plt.show()
```

```python
print(fBm_chi_square_test(h05, 0.5, 0.95))
print(fBm_chi_square_test(h01, 0.1, 0.95))
```

```python
size = 1000
rho = 0.6
bfbm_gen = naive.NaiveBiFBmGenerator()
bfbm = bfbm_gen.generate_bifBm(H1=0.5, H2=0.1, rho=rho, size=size)

plt.plot(range(size), bfbm[0], label='H=0.5')
plt.plot(range(size), bfbm[1], label='H=0.1')
plt.legend()
plt.show()
```

```python
bfgn = np.diff(bfbm)
plt.plot(bfgn[0], bfgn[1], 'o')
plt.plot(np.arange(-4, 4, 0.01), rho*np.arange(-4, 4, 0.01))
plt.show()
```
