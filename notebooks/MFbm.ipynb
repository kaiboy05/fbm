```python
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
```

```python
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

from fbm.sim import naive
bfBm = naive.NaiveBiFBmGenerator()
```

```python
size = 999
H1 = 0.5
H2 = 0.5
rho = 0.6

ts = bfBm.generate_bifBm(H1, H2, rho, size)
plt.plot(range(size), ts[0], label=f"H1={H1}")
plt.plot(range(size), ts[1], label=f"H2={H2}")
plt.title(f"Rho = {rho}")
plt.legend()
plt.show()

rho = -0.3

ts = bfBm.generate_bifBm(H1, H2, rho, size)
plt.plot(range(size), ts[0], label=f"H1={H1}")
plt.plot(range(size), ts[1], label=f"H2={H2}")
plt.title(f"Rho = {rho}")
plt.legend()
plt.show()
```
